#### Styling tabulation

```python
df = df['2021-04-27 00:00:00':'2021-05-03 00:00:00']
```

```python
df.style
```

```python
def func(x):
    color='red' if x>10000 else 'blue'
    return 'color:'+color
```

Displays values above 10000 in red and rest in blue.

```python
df.style.applymap(func)
```

Highlights the max value of the column in orange

```python
df.style.highlight_max(color='orange')
```

```python
df.drop(['dd','ld'],axis=1,inplace=True)
```

Highlights the max value of the column in orange and minimum in green

```python
df.style.highlight_max(color='orange').highlight_min(color='green')
```

```python
df.drop('tt',axis=1,inplace=True)
```

```python
def bold_max_value(x):
    is_max= (x==x.max())
    return ['font-weight:bold' if y else '' for y in is_max]
```

Bolds max value and highlights min value in green across each col

```python
df.style.apply(bold_max_value).highlight_min(color='green')  
```

Bold for col wise max and red for row wise max

```python
df.style.apply(bold_max_value).highlight_max(color='red',axis=1) 
```

Column wise heat map

```python
df.style.background_gradient(cmap='Reds') 
```

Row wise heat map

```python
df.style.background_gradient(cmap='Reds',axis=1)  
```

Subset of row wise heat map

```python
df.style.background_gradient(cmap='Reds',subset=['mh','dl','tn']) 
```

```python
df.style.bar()
```

Column wise bar plots

```python
df.style.bar(subset=['mh','dl','tn'])
```

```python
df[['mh','dl','tn']].style.bar()
```

```python
df[['mh','dl','tn']].style.bar(subset=['mh'],color='red').bar(subset=['tn'],color='yellow').bar(subset=['dl'],color='blue')
```
